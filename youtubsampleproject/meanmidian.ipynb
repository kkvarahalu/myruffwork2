```python
df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
sns.boxplot(df['Fare'])
```

```python
df['Fare']=df['Fare'].fillna(df['Fare'].median())
```

```python
df['Age'].describe()
```

```python
q1=df['Age'].describe()['25%']
q3=df['Age'].describe()['75%']
iqr=q3-q1
lb=q1-1.5*iqr
ub=q3+1.5*iqr
df['Age']=df['Age'].clip(lb,ub)
```

```python
sns.boxplot(df['Age'])
```

```python
q1=df['Fare'].describe()['25%']
q3=df['Fare'].describe()['75%']
iqr=q3-q1
lb=q1-1.5*iqr
ub=q3+1.5*iqr
df['Fare']=df['Fare'].clip(lb,ub)
```

```python
sns.boxplot(df['Fare'])
```
