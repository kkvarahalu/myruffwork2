```python
dataset=pd.read_csv('Churn_Modelling (1).csv')
```

```python
x=dataset.iloc[:, 3:-1].values
y=dataset.iloc[:, -1].values
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
x[:, 2]=le.fit_transform(x[:, 2])
```

```python
x
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(), [1])],remainder='passthrough')
x=np.array(ct.fit_transform(x))
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### ANN MODEL BUILDING ARCHITECTURE


```python
ann=tf.keras.models.Sequential() #establish layer one ofter another
```

```python
#layer connected to nuren called -dense
```

```python
ann.add(tf.keras.layers.Dense(units=6))
```

```python
ann.add(tf.keras.layers.Dense(units=6,activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=6,activation='sigmoid'))
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann = tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units=6, activation='relu', input_shape=(x_train.shape[1],)))
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))  # <-- fixed output layer

# Compile
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
ann.fit(x_train,y_train,batch_size=32,epochs=10)
```

```python
#we try to 
#the loss is reduce then acuracy is reduce
y_pred=ann.predict(x_test)
y_pred=(y_pred>0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
acc=accuracy_score(y_test,y_pred)
print("Confusion Matrix:\n",cm)
print("Accuracy:",acc)
```
