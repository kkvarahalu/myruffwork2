```python
e.rename(columns={'id':'match_id'},inplace=True)
```

```python
merged=pd.merge(e,f,on='match_id',how='inner')
```

```python
merged
```

```python
#get all deliveries where the team1 from the matches dataset is "Mumbai Indians"?
result=merged[merged['team1']=='Mumbai Indians']
```

```python
result
```

```python
#batting_team is the same as team1 from the matches dataset?
result=merged[merged["batting_team"]==merged["team1"]]
```

```python
result
```

```python
#the player_of_match column has the same value as the batter column from deliveries?
a=merged[merged["player_of_match"]==merged["batter"]]
```

```python
a
```

```python
merged.columns
```

```python
result
```

```python
#filter the results to only show matches played in a specific city (e.g., "Mumbai")?
result=merged[merged['city']=='Mumbai']
```

```python
result
```

```python
def match(city):
    result=merged[merged['city']==city]
    return result
```

```python
match('Bangalore')
```

```python
def match(city):
    result=merged[merged['city']==city]
    return result.shape[0]
```

```python
match('Mumbai')
```

```python
#calculate the total runs scored by each batting_team in each match?
result=merged.groupby(["match_id","batting_team"])["total_runs"].sum().reset_index()
```

```python
result
```

```python
#get the total number of wickets (is_wicket) taken by each bowling team in each match?
result=merged.groupby(["match_id","bowling_team"])["is_wicket"].sum().reset_index()
```

```python
result
```

```python
l_matches=merged[merged["match_type"]=="League"]
result=l_matches.groupby("batter")["total_runs"].sum()
```

```python
result
```

```python
#find out which venues are associated with specific batting_team and bowling_team pairs?
result=merged.groupby(["batting_team","bowling_team"])["venue"].unique().reset_index()
```

```python
result
```
