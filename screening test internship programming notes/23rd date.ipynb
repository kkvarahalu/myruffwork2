```python
#binary search tree;
#Binary search:element should be in a sorted order
# element is not in lst thet is worest case in binary tree
# lst will contain sorted order
def binarysertree(lst,low,high,key):
    if lst[low]>lst[high]:
        return False
    mid=(low+high)//2
    if lst[mid]==key:
        return True
    elif key<lst[mid]:
        return binarysertree(lst,low,mid-1,key)
    elif lst[mid]<key:
        return binarysertree(lst,mid+1,high,key)
    else:
        return 0
lst=[5,3,7,2,4,6,8] 
lst=sorted(lst)
print(binarysertree(lst,0,len(lst)-1,5))
```

```python
def binarysertree(lst,low,high,key):
    if low>high:
        return False
    mid=(low+high)//2
    if lst[mid]==key:
        return True
    elif key<lst[mid]:
        return binarysertree(lst,low,mid-1,key)
    elif lst[mid]<key:
        return binarysertree(lst,mid+1,high,key)
    else:
        return 0
def sor(lst):
    for i in range(len(lst)):
        for j in range(i+1,len(lst)):
            if lst[i]>lst[j]:
                lst[i],lst[j]=lst[j],lst[i]
    return lst
lst=[3,4,5,6,7,8]
lst=sor(lst)
print(binarysertree(lst,0,len(lst)-1,5))
```

```python
def recursive_my_count(lst,element):
    if lst==[]:
        return 0
    elif lst[0]==element:
        return 1+recursive_my_count(lst[1:],element)
    else:
        return recursive_my_count(lst[1:],element)
lst=[2,2,3,4,5,6,7,7,9,15]
print(recursive_my_count(lst,7))
```

```python
#make one change
def greatest_from_left(lst):
    result=[-1]
    max1=lst[0]
    for i in range(1,len(lst)):
        max1=lst[i]
        
        for j in range(0,i+1):
           
            if lst[j]>max1:
                max1=lst[j]
        if max1>lst[i]:
            result.append(max1)
        else:
            result.append(-1)
        
    return result
lst=[70,30,50,45,50,46,78,67,70,75,100]
print(greatest_from_left(lst))
```

```python
def gret_form_left(lst,f):
    r=[]
    for i in range(len(lst)):
        if i==0:
            r.append(f)
        else:
            for j in range(i-1,-1,-1):
                if lst[i]<lst[j]:
                    r.append(lst[j])
                    break
            else:
                r.append(f)
    return r
lst=[70,30,50,45,50,46,78,67,70,75,100]
print(gret_form_left(lst,-1))
```

```python
def valid_email(email):
    pattern='[a-zA-Z]+[0-9\_\.a-z]+\@[a-z]+\.[a-z]+'
    if re.findall(pattern,email):
        return True
    else:
        return False
print(valid_email("mayur.kulkarni636@gmail.ac.in"))
```
