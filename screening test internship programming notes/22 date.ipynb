```python
def all_palin(s):
    for i in range(len(s)):
        for j in range(i+1,len(s)+1):
            ss=s[i:j]
            if ispallindrome(ss):
                print(ss)
def ispallindrome(s):
    i=0
    j=len(s)-1
    while i<j:
        if s[i]==s[j]:
            i+=1
            j-=1
        else:
            return False
    return True
s="abababababababababbbbbbababababab"
print(all_palin(s))
```

```python
def palinnum(s):
    l=[]
    for i in range(len(s)):
        for j in range(i+1,len(s)+1):
            subst=s[i:j]
            if subst==subst[::-1]:
                l.append(subst)
    return l
s="abababababababababbbbbbababababab"
print(palinnum(s))
```

```python
def compress_substr(s):
    prev=0
    net=prev+1
    ss='a'
    while net<len(s):
        if s[prev]!=s[net]:
            ss+=s[net]
        prev=net
        net=prev+1
    print(ss)
s="aaabbbbaaaacccc"
compress_substr(s)
```

```python
def addstnum(s):
    d={}
    ans=''
    for i in s:
        if i in d:
            d[i]+=1
        else:
            d[i]=1
    for k,v in d.items():
        ans+=k+str(v)
    return ans
s='aaaabbbbcccccc'
print(addstnum(s))
```

```python
# search: Scans through the string and returns the first match object.
# If no match is found â†’ returns None.
# You can use .group() or .groups() on the match object to see the matched parts.

import re
s="mayurkularni636@gmail.com"
p='([a-zA-Z0-9]+)(\@[a-zA-Z0-9]+)(\.)([a-zA-Z0-9]+)'
match=re.search(p,s)
```

```python
match.group(1)
```

```python
match.group(2)
```

```python
match.group(3)
```

```python
match.group(4)
```
