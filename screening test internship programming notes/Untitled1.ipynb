```python
class facebook_friend_recommendation:
    def __init__(self):
        self.graph={}
    def add_friend(self,p1,p2):
        if self.graph=={}:
            self.graph[p1]=[p2]
            self.graph[p2]=[p1]
        else:
            if p1 in self.graph:
                if p2 not in self.graph[p1]:
                    self.graph[p1]=p2
```
