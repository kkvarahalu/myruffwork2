```python
def min_max(lst,low,high):
    if low==high:
        return lst[low],lst[high]
    if high==low+1:
        if lst[low]<lst[high]:
            return lst[low],lst[high]
        else:
            return lst[high],lst[low]
    mid=(low+high)//2
    left_min,left_max=min_max(lst,low,mid)
    right_min,right_max=min_max(lst,mid+1,high)
    return min(left_min,right_min),max(left_max,right_max)
lst=[7, 2, 9, 4, 1, 5]
print(min_max(lst,0,len(lst)-1))
```
