```python
#Find all prime numbers between 1–100.
def prm(n):
    if n<=1:
        return False
    elif n==2:
        return True
    for i in range(2,n):
        if n%i==0:
            return False
        else:
            return True
n=100
for j in range(2,n):
    if prm(j):
        print(j)
```

```python
#Count frequency of each element in a list.
def count(n):
    d={}
    for i in n:
        if i in d:
            d[i]+=1
        else:
            d[i]=1
    return d
n=[1,2,2,3,3,4,4,4,5,6]
count(n)
```

```python
# Write a function to check if a number is Armstrong.
#ex=153=1+125+27
def armstro(n,i,r):
    while n>0:
        l=n%10
        r=r+l**3
        n=n//10
    return r
armstro(153,3,0)
```
