```python
def count_no_set_bit(n):
    c=0
    while n>0:
        if n&1!=0:
            c+=1
        n=n>>1
    return c
count_no_set_bit(12)
```

```python
#integer is stored in 32 bits
def count_no_set_bit(n):
    c=0
    for i in range(1000):
        if i&1!=0:
            c+=1
        n=n>>1
        if n==0:
            break
    return c
count_no_set_bit(12)
```

```python
# check kth-bit 
# 3. Check if k-th bit (from right) is set (Input: n = 12 (1100), k = 2)  # Output: True
n=12
k=2
if n&(1<<k)!=0:
    print(True)
else:
    print(False)
```

```python
# swap number in bitwise operater
def swap(a,b):
    a=a^b
    b=a^b
    a=a^b
    return a,b
print(swap(4,5))
```
