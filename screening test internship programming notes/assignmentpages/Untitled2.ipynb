```python
#compute the area under f(x)=cos(x) between o and pi
import math
a=0
b=math.pi
n=100000
dx=(b-a)/n
area=0
x=a
for i in range(n):
    area+=math.cos(x)*dx
    x+=dx
print(area)
```

```python
#find the a point is inside or outside a rectanngle(code)
def checkins(x,y,x1,y1,x2,y2):
    return x1<=x<=x2 and y1<=y<=y2
checkins(5,4,2,3,8,9)
```

```python
#find if two circles intersect at exactly one point or not(code)
def check_int(x1,y1,r1,x2,y2,r2):
    d=((x2-x1)**2 + (y2-y1)**2)**0.5
    return d==r1+r2 or d==abs(r1-r2)
check_int(3,4,5,6,7,8) 
```
