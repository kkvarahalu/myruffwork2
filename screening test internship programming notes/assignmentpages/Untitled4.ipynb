```python
# Q1.Write a function that generates a specified number of independent and identically distributed (i.i.d.) draws from a
# uniform distribution between 0 and 1. The function should take the number of draws as an argument and return a list of random numbers.
# Note: Use a fixed random seed of 42 to ensure that the results are reproducible.
# Test Case #1
# Input: 5
# Output: "[0.37, 0.95, 0.73, 0.6, 0.16]"
# Description: This test case assesses the function"s correctness in generating a small number of i.i.d. draws from the uniform 
# distribution between 0 and 1.
# Test Case #2
# Input: 100
# Output: "A list of 100 random numbers between 0 and 1, e.g., [0.134, 0.245, 0.756, ..., 0.890]. "
# Description: This test validates the function"s ability to handle a larger number of i.i.d. draws, ensuring the output
# remains within the expected range and maintains the properties of the distribution.
# Test Case #3
# Input: 0
# Output: "An empty list, []."
# Description: This case challenges the function"s handling of the edge case where no draws are requested, ensuring it 
# returns an empty list appropriately.
import numpy as np
def uniform(n):
    np.random.seed(42)
    r=np.random.uniform(0,1,n)
    return list(np.round(r,2))
print(f'first qus ans:{uniform(5)}')
print(f'3rd qus ans:{uniform(0)}')
```

```python
# Q5 Write a function that samples the flip of a biased coin for N times.
# Test Case #1
# Input: {"N": 10, "probability": 0.5}
# Output: [1, 1, 0, 0, 1, 0, 1, 0, 0, 1]
# Description: This test case checks the function with a fair coin flip (probability 0.5) for 10 trials.
# Test Case #2
# Input: {"N": 5, "probability": 0.8}
# Output: [1, 1, 1, 1, 1]
# Description: This test case uses a biased coin with a higher probability of heads (0.8) for 5 trials.
# Test Case #3
# Input: {"N": 15, "probability": 0.1}
import random
d={"N": 10, "probability": 0.5}
ans=[]
for i in range(d['N']):
    if random.random()<d["probability"]:
        ans.append(1)
    else:
        ans.append(0)
print(ans)
```
