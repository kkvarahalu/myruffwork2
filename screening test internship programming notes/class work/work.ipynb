```python
import numpy as np
import random
def permutation_test(s1,s2):
    D=abs(np.mean(s1)-np.mean(s2))
    combined_data=s1+s2
    l1=len(s1)
    l2=len(s2)
    no_of_samples=1000
    count=0
    new_s1=[]
    new_s2=[]
    sum=0
    result=[]
    for i in range(no_of_samples):
        rs=random.sample(combined_data,len(combined_data))
        for i in range(l1):
            new_s1.append(rs[i])
        for j in range(l1,l1+49):
            new_s2.append(rs[j])
        for i in range(len(new_s1)):
            sum+=new_s1[i]
        mean1=sum/len(new_s1)
        for i in range(len(new_s2)):
            sum+=new_s2[i]
        mean2=sum/len(new_s2)
        result.append(abs(mean2-mean1))
    result.sort()
    for i in range(len(result)):
        if D>result[i]:
            count+=1
    print("p_value:",count/len(result))
```

```python
s1=[80,90,75,85,95]
s2=[85,88,78,80,90]
permutation_test(s1,s2)
```
