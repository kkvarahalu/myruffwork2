```python
#1
def f(n):
    if n<=1:
        return n
    if n%2==0:
        return f(n//2)
    return f(n//2)+f(n//2+1)
print(f(11))
```

```python
# 2
def f(n):
    i=[1]
    if n>=5:
        return n
    n=n+i[0]
    i[0]+=1
    return f(n)
print(f(1))
```

```python
# 3
def fun(n):
    if n>0:
        fun(n-1)
        print(n-1,end='') 
        fun(n-1)
fun(3)
```

```python
# 4
def fun1(n):
    if n==0:
        return 
    print(n,end='') # printing without a newline
    fun2(n-2)
    print(n,end='')
def fun2(n):
    if n==0:
        return
    print(n,end='')
    fun1(n+1)
    print(n,end='')
fun1(3)
```

```python
# 5
def func(n):
    if n<3:
        return 1
    else:
        return func(n-1)+func(n-3)+1
print(func(5))
```
