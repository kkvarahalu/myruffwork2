```python
# 18. Check if a Set is a Superset
# Write a Python program to check if a given set is a superset of itself and a
# superset of another given set.
a={1,2,3,4,5}
b={2,3}
if a.issuperset(a):
    print("a is a superset of itself") 
if a.issuperset(b):
    print("a is a superset of b")
else:
    print("a is not a superset of b")
```
