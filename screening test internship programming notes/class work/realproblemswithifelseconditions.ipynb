```python
# 1. Equality Check
# Write a Python code to accept two integers and check whether they are equal
# or not.
# Test Data : 15 15
# Expected Output :
# Number1 and Number2 are equal

# method1 normal
#simple:
# def two_num(n1,n2):
#     if n1==n2:
#         print('Number1 and Number2 are equal')
#     else:
#         print('false')
# two_num(15,15)

# method 2 bitwise :
# def two_num(n1,n2):
#     if n1^n2==0:
#         print('Number1 and Number2 are equal')
#     else:
#         print('false')
# two_num(15,15)

# method 3 recursion:
def two_num(n1,n2):
    if n1==0 and n2==0:
        print('Number1 and Number2 are equal')
    elif n1==0 or n2==0:
        return False
    elif n1<0 and n2<0:
        return two_num(-n1,-n2)
    elif n1<0 or n2<0:
        return False
    elif n1<0 and n2>0 or n2<0 and n1>0:
        return False
    else:
        return two_num(n1-1,n2-1)
two_num(15,15)
```

```python
# 2. Even or Odd Check
# Write a Python code to check whether a given number is even or odd.
# Test Data : 15
# Expected Output :
# 15 is an odd integer

#method 1:
# def even_odd(n):
#     if (n&1)==0:
#         return 'even'
#     elif n&1!=0:
#         return 'odd'
#     else:
#         return False
# even_odd(15)

# method 2  recurtion:
def even_odd(n):
    if n==0:
        return 'even'
    elif n==1:
        return  'odd'
    elif n<0:
        return even_odd(-n)
    else:
        return even_odd(n-2)
even_odd(15)
```

```python
# 4. Leap Year Determination
# Write a Python code to find whether a given year is a leap year or not.
# Test Data : 2016
# Expected Output :
# 2016 is a leap year.
def leap_year(n):
    if (n % 400 == 0) or (n % 4 == 0 and n % 100 != 0):
        return 'Yes'
    else:
        return 'No'

print(leap_year(1996))
```

```python
# 13. Temperature-Based Weather Message
# Write a Python code to read temperature in centigrade and display a suitable message according to the temperature state below:
# Temp < 0 then Freezing weather
# Temp 0-10 then Very Cold weather
# Temp 10-20 then Cold weather
# Temp 20-30 then Normal in Temp
# Temp 30-40 then Its Hot
# Temp >=40 then Its Very Hot
# Test Data :
# 42
# Expected Output :
# Its very hot.
def temp(data):
    if data<0:
        print('Freezing weather')
    elif data>=0 and data<=10:
        print("Very Cold weather")
    elif data>10 and data<=20:
        print('Cold weather')
    elif data>20 and data<=30:
        print('Normal in Temp')
    elif data>30 and data<40:
        print('Its Hot')
    else:
        print('Its Very Hot')
temp(42)
```
