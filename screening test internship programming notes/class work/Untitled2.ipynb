```python
def su(n):
    if n==0:
        return 0
    return (n%10)+su(n//10)
su(19)
```

```python
def rev(n):
   rev=0
   r=n
   while r>0:
      last_digit=r%10
      rev=rev*10+last_digit
      r=r//10
   print(rev)
rev(323)
```

```python
def reversenum(x):
    if x<0:
        return -int(str(abs(x))[::-1])
    else:
        return int(str(x)[::-1])
print(reversenum(-123))
print(reversenum(123))
```
