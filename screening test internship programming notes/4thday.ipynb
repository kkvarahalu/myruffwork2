```python
def min_max(lst,low,high):
    if low==high:
        return lst[low],lst[high]
    if high==low+1:
        if lst[low]<lst[high]:
            return lst[low],lst[high]
        else:
            return lst[high],lst[low]
    mid=(low+high)//2
    left_min,left_max=min_max(lst,low,mid)
    right_min,right_max=min_max(lst,mid+1,high)
    minvalue=smaller(left_min,right_min)
    maxvalue=greater(right_max,right_max)
    return minvalue,maxvalue
def smaller(a,b):
    if a<b:
        return a
    else:
        return b
def greater(x,y):
    if x<y:
        return x
    else:
        return y
lst=[2,-1,-3,7,9,19,21,33,81,25]
low=0
high=len(lst)-1
print(min_max(lst,low,high))
```

```python
def mergin(lst1,lst2):
    i=j=0
    result=[]
    while i<len(lst1) and j<len(lst2):
        
        if lst1[i]<lst2[j]:
            result.append(lst1[i])
            i+=1
        else:
            result.append(lst2[j])
            j+=1
    while i<len(lst1):
        result.append(lst1[i])
        i+=1
    while j<len(lst2):
        result.append(lst2[j])
        j+=1
    return result
lst1=[4,2,3,4]
lst2=[3,4,6,7]
print(mergin(lst1,lst2))
```

```python
# peak value find 
# ex=[1,2,-1]--> lst[1]>lst[0] and lst[1]>lst[2] so it is peak value
def peakvs(lst):
    if lst[0]>lst[1]:
        print(lst[0])
    if lst[-1]>lst[-2]:
        print(lst[-1])
    for i in range(1,len(lst)-1):
        if lst[i]>lst[i-1] and lst[i]>lst[i+1]:
            print(lst[i])
lst=[12,2,3,4,-2,56,32,45,22,81]
print(peakvs(lst))
```
