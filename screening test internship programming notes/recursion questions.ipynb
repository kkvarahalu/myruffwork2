```python
#1.Factorial of a number: Write a recursive function to calculate the factorial of a number n (i.e., n!).
def facto(n):
    if n==0 or n==1:
        return 1
    else:
        return n*facto(n-1)
facto(5)
```

```python
#2Fibonacci series: Write a recursive function to generate the n-th number of the Fibonacci sequence.
def fibo(n,a=0,b=1):
    if n==0:
        return a
    elif n==1:
        return b
    else:
        return fibo(n-1,b,a+b)
n=10
for i in range(n):
    print(fibo(i))
```

```python
# 3. Sum of first n natural numbers: Write a recursive function to calculate
# the sum of the first n natural numbers.
def naturalnu(n):
    if n<=0:
        return 'not natural numbers'
    if n==1:
        return n
    else:
        return n+naturalnu(n-1)
print(naturalnu(-5))
```

```python
# 4. Power of a number: Write a recursive function to calculate x^n (x raised to the power of n). 4**2=16
def powern(n,a):
    if a==0:
        return 1
    else:
        return n*powern(n,a-1)
print(powern(4,3))
```

```python
def powern(n,a):
    if a==0:
        return 1
    elif a%2==0:
        res=powern(n,a//2)
        return res*res
    else:
        res=powern(n,a//2)
        return res*res*n
print(powern(4,4))
```

```python
# 5. Reverse a string: Write a recursive function to reverse a string.
def rev(s):
    if s=='':
        return s
    else:
        return rev(s[1:])+s[0]
s='abcdf'
print(rev(s))
```

```python
# 7. Count occurrences of a character in a string: Write a recursive function
# to count how many times a particular character appears in a string.
def coun_chr(s,ch):
    if s=='':
        return 0
    elif s[0]==ch:
        return 1+coun_chr(s[1:],ch)
    else:
        return coun_chr(s[1:],ch)
s='kkvarahalu'
ch='k'
print(coun_chr(s,ch))
```

```python
# 12. Find the minimum element in an array: Write a recursive function to
# find the smallest element in an array.
def minv(lst,low,high):
    if low==high:
        return lst[low]
    elif high==low+1:
        if lst[low]<lst[high]:
            return lst[low]
        else:
            return lst[high]
    mid=(low+high)//2
    left_min=minv(lst,low,mid)
    right_min=minv(lst,mid+1,high)
    minvalue=small(left_min,right_min)
    return minvalue
def small(a,b):
    if a<b:
        return a
    else:
        return b
lst=[3,2,1,5,4,7]
print(minv(lst,0,len(lst)-1))
```

```python
# 13. Find the maximum element in an array: Write a recursive function to
# find the largest element in an array.
def maxv(lst,l,h):
    if l==h:
        return lst[h]
    elif h==l+1:
        if lst[l]<lst[h]:
            return lst[h]
        else:
            return lst[l]
    mid=(l+h)//2
    left_max=maxv(lst,l,mid)
    right_max=maxv(lst,mid+1,h)
    max_value=begger(left_max,right_max)
    return max_value
def begger(a,b):
    if a<b:
        return b
    else:
        return a
lst=[3,2,1,5,9,4,7]
print(maxv(lst,0,len(lst)-1))
```

```python
# 14 Find the length of a string: Write a recursive function to find the length
# of a string.
def lenstr(s):
    if s=='':
        return 0
    else:
        return 1+lenstr(s[1::])
s='abc'
print(lenstr(s))
```

```python
#17. Check if a string is a palindrome: Write a recursive function to check if a string is a palindrome.
def is_palin(s):
    if s=='':
        return 'palindrome'
    elif len(s)==1:
        return 'palindrome'
    elif s!=s[::-1]:
        return 'not palindrome'
    else:
        return is_palin(s[1:-1])
s='madam'
print(is_palin(s)) 
```
