```python
def add_two_nu(n,a):
    if a==0:
        return n
    elif a<0 and n<0:
        return -add_two_nu(-n,-a)
    elif a<0:
        return add_two_nu(a+1,n-1)
    elif n<0:
        return add_two_nu(a-1,n+1)
    else:
        r=add_two_nu(n+1,a-1) #add reverse in stock =3+1+1=5 because i will give if a==0 then return n so add 3 like n value(next line in belowe)
        return r
print(add_two_nu(8,0))
print(add_two_nu(6,2))
print(add_two_nu(100,500))
```

```python
# wap division using recursion?
# first i try without negitive
def divid(a,b):
    if a==0:
        return 0
    elif b==0:
        return 'Value error'
    elif a<0 and b<0:
        return divid(-a,-b)
    elif a<0:
        return -divid(-a,b)
    elif b<0:
        return -divid(a,-b)
    else:
        return 1+divid(a-b,b)
print(divid(7,2))
```

```python
def divid(a,b):
    if b==0:
        return "Value error"
    if a<0 and b<0:
        return divid(-a,-b)
    elif a<0:
        return -divid(-a,b)
    elif b<0:
        return -divid(a,-b)
    if a==b:
        return 0
    if a<b:
        def decimal(a,b):
            if a==0:
                return 0
            elif a<b:
                def dec_decimal(a,b):
                    if a==0:
                        return 0
                    elif a<b:
                        return 0
                    else:
                        return 0.01+dec_decimal(a-b,b)
                return dec_decimal(a*10,b)
            else:
                return 0.1+decimal(a-b,b)

        return decimal(a*10,b)
    return 1+divid(a-b,b)
print(divid(2,4))
```

```python
# divide an conquer strategy:
def power(a,b):
    if b==0:
        return 1
    elif b%2==0:
        res=power(a,b//2)
        return res*res
    else:
        res=power(a,b//2)
        return res*res*a
power(7,2)
```

```python
# factorial numbers in recursive approach
def fac(n):
    if n==0:
        return 1
    else:
        return n*fac(n-1)
print(fac(5))
```

```python
# python fibnumber in recursion
def fib(n):
    if n<=1:
        return n
    else:
        return (fib(n-1)+fib(n-2))
n=5
for i in range(n):
    print(fib(i))
```

```python
def fibi(s,a=0,b=1):
    if s==0:
        return a
    elif s==1:
        return b
    else:
        return fibi(s-1,b,a+b)
s=1
for i in range(s):
    print(fibi(i))
fibi(s)
```

```python
def fibn(inp,a=0,b=1):
    if inp==0:
        return a
    elif inp==1:
        return b
    else:
        return fibn(inp-1,b,a+b)
inp=10
fib_sum=[fibn(i) for i in range(inp)]
print(fib_sum)
print(sum(fib_sum))
```

```python
#reverse string
def rev(s):
    if len(s)==1:
        return s
    else:
        return rev(s[1:])+s[0]
s='kkvarahalu'
print(rev(s))
```

```python
def sub(a,b):
    if b==0:
        return a
    elif b>0 :
        return sub(a-1,b-1)
    elif a<0 and b<0:
        return -sub(-a,-b)
    elif b<0:
        return sub(a-1,b+1)
    else:
        return sub(a+1,b+1)
print(sub(3,4))
print(sub(5,-3))
print(sub(-5,-3))
print(sub(5,-3))
print(sub(-2,-3))
```

```python
#len lst size
def len_lst(lst):
    if lst==[]:
        return 0
    else:
        return 1+len_lst(lst[1::]) 
lst=[1,2,3,4,5]
print(len_lst(lst))
 # stack=5,4,3,2,1=1+
 #   =4,3,2,1=1+
 #     =3,2,1=1+
 #     =2,1=1+
 #     =1=1+
 #     =0=0+1+1+1+1+1=5....output      
```
