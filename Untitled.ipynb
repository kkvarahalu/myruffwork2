```python
b=pd.read_csv("matches.csv")
```

```python
c = pd.merge(b,a, left_on="id", right_on="match_id", how="inner")
```

```python
c
```

```python
def matches_city(city):
    result=c[c["city"]==city]
    return result
```

```python
matches_city("Chennai")
```

```python
def matches_citycount(city):
    result=c[c["city"]==city]
    return result.shape[0]
```

```python
matches_citycount("Chennai")
```
