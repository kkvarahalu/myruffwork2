```python
#removel of html tags
def remove_tags(text):
    pattern=re.compile('<.*>')
    return pattern.sub(r'',text)
```

```python
print(remove_tags(text))
```

```python
# remove punctuations
import string
string.punctuation
```

```python
def remove_punctuation(text):
    result = ''
    for char in text:
        if char not in string.punctuation:
            result += char
    return result
df['no_punctuation'] = df['message'].apply(remove_punctuation)
```
