```python
def ind(l,i,v):
    r=[]
    for j in range(len(l)):
        if j==i:
            r.append(v)
        r.append(l[j])
    return r
l=[1,2,3,4,5,6,7]
ind(l,2,9)
```

```python
def ind(l,i,v): 
    r=[]
    for j in range(0,i):
        r+=[l[j]]
    r+=[v]
    r+=l[i:]
    return r
l=[1,2,3,4,5,6,7]
ind(l,2,9)
```
