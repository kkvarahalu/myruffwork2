```python
# 6. Generate Dictionary of Numbers and Their Squares
# Write a Python script to generate and print a dictionary that contains a number
# (between 1 and n) in the form (x, x*x).
# Sample Dictionary ( n = 5) :
# Expected Output : {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
n=int(input())
l=[]
for i in range(1,n):
    l.append(i)
d={}
for i in l:
     if i in d:
        d[i]+=i**2
     else:
        d[i]=i**2
print(d)
```

```python
def min_v(l,low,high):
    if low==high:
        return l[low]
    elif high==low+1:
        if l[low]<l[high]:
            return l[low]
        else:
            return l[high]
    mid=(low+high)//2
    left_min=min_v(l,low,mid)
    right_min=min_v(l,mid+1,high)
    return small(left_min,right_min)
def small(a,b):
    if a<b:
        return a
    else:
        return b
l=[6,7,1,2,3,4,5]
print(min_v(l,0,len(l)-1))
```

```python
def binary(l,low,high,key):
    if l[low]>l[high]:
        return False
    mid=(low+high)//2
    if l[mid]==key:
        return True
    elif l[mid]<key:
        return binary(l,mid+1,high,key)
    elif l[mid]>key:
        return binary(l,low,mid-1,key)
    else:
        return 0
def sortv(l):
    return sorted(l)
l=sortv([5,3,7,2,4,6,8])
print(binary(l,0,len(l)-1,8))
```
