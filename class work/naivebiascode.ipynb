```python
import pandas as pd
```

```python
data={
    'Weather': ['Sunny','Sunny','Overcast','Rain','Rain','Rain','Overcast','Sunny','Sunny','Rain','Sunny','Overcast','Overcast','Rain'],
    'Temperature': ['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild','Cool','Mild','Mild','Mild','Hot','Mild'],
    'Play': ['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
p_yes=(df['Play']=='Yes').sum()/len(df)
p_no=(df['Play']=='No').sum()/len(df)
v=len(df['Weather'].unique())
rain_yes=(df[(df['Weather']=='Rain')&(df['Play']=='Yes')].shape[0]+1)/((df['Play']=='Yes').sum()+v)
rain_no=(df[(df['Weather']=='Rain')&(df['Play']=='No')].shape[0]+1)/((df['Play']=='No').sum()+v)
mild_yes=(df[(df['Temperature']=='Mild')&(df['Play']=='Yes')].shape[0]+1)/((df['Play']=='Yes').sum()+v)
mild_no=(df[(df['Temperature']=='Mild')&(df['Play']=='No')].shape[0]+1)/((df['Play']=='No').sum()+v)
```

```python
P_yes=p_yes*rain_yes*mild_yes
P_no =p_no*rain_no*mild_no
print("P(Yes):",P_yes)
print("P(No):",P_no)
if P_yes>P_no:
    print("Decision:","Yes")
else:
     print("Decision:","No")
```
