```python
# Q3. Problem -Statement –
# Joseph is learning digital logic subject which will be for his next semester. 
# He usually tries to solve unit assignment problems before the lecture. Today he got one tricky question.
# The problem statement is “A positive integer has been given as an input. Convert decimal value to binary representation.
# Toggle all bits of it after the most significant bit including the most significant bit. 
# Print the positive integer value after toggling all bits”.
# Constraints-
# 1<=N<=100
# Example 1:
# Input :
# 10  -> Integer
# Output :
# 5    -> result- Integer
# Explanation:
# Binary representation of 10 is 1010. After toggling the bits(1010), will get 0101 which represents “5”. Hence output will print “5”.

def bit_tog(n):
    c=0
    t=n
    while t>0:
        t//=2
        c+=1
    m=(1<<c)-1
    return n^m
bit_tog(10) 
```

```python
# Q4. Problem: The Mystery Number Sequence
# A mysterious sequence of numbers is formed by a pattern. The sequence starts with the number 1, and each subsequent 
# number is generated based on the previous number, but with a twist. The sequence grows based on two specific rules:
# 1.If the number is odd, the next number is number * 3 + 1.
# 2.If the number is even, the next number is number / 2.
# The sequence continues until the number reaches 1.
# Your Task: Write a program that generates the sequence starting from a given number n, and 
# returns the sequence of numbers until it reaches 1.
# Example:
# Input: n = 6
# Output:
# 6, 3, 10, 5, 16, 8, 4, 2, 1
def holist(n):
    while True:
        print(n)
        if n==1:
            break
        elif n%2!=0:
            n=n*3+1
        elif n%2==0:
            n=n//2
holist(6)
```

```python
#Q5. Problem Statement: Given a number N, print the smallest and largest digits present in the number.
# Example 1:
# Input: N = 2746
# Output: Largest digit: 7
#         Smallest digit: 2
# Explanation: By simply going through the digits of 
# the number, we figure out the largest and smallest 
# digit in the number.

# Example 2:
# Input: N = 23004
# Output: Largest digit : 4
#         Smallest digit : 0
# Explanation: By simply going through the digits of 
# the number, we figure out the largest and smallest digit in the number.
n=2746
r=0
def rotate_number(n,r):
    ans=[]
    while n>0:
        l=n%10
        ans.append(l)
        r=r*10+l
        n=n//10
    result=sorted(ans)
    return result[-1],result[0]
print(rotate_number(n,r))
```

```python
# Q6. Problem Statement: Check if the given year is a leap year or not.
# Example 1:
# Input: 1996
# Output: Yes
# Explanation: Since 1996 is a leap year answer is “Yes”.

# Example 2:
# Input: 2000
# Output: Yes
def leap_year(n):
    if (n%400==0) or (n%4==0 and n%100!=0):
        return 'Yes'
    else:
        return 'No'
print(leap_year(1996))
print(leap_year(2000))
print(leap_year(1900))
print(leap_year(100))
```

```python
# Q7. Problem Statement: Given a and b, find prime numbers in a given range [a,b], (a and b are included here).
# Examples:
# Input: 2 10
# Output: 2 3 5 7 
# Explanation: Prime Numbers b/w 2 and 10 are 2,3,5 and 7.

# Example 2:
# Input: 10 16
# Output: 11 13 
# Explanation: Prime Numbers b/w 10 and 16 are 11 and 13.
def pime_no(n):
    if n==1:
        return False
    elif n==2:
        return True
    for i in range(2,n):
        if n%i==0:
            return False
    else:
        return True
n1=int(input())
n2=int(input())
for j in range(n1,n2+1):
    if pime_no(j):
        print(j,end=' ')
```
