```python
# Q1. Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.
# You may assume that each input would have exactly one solution, and you may not use the same element twice.
# You can return the answer in any order.
 
# Example 1:
# Input: nums = [2,7,11,15], target = 9
# Output: [0,1]
# Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].
# Example 2:
# Input: nums = [3,2,4], target = 6
# Output: [1,2]
# Example 3:
# Input: nums = [3,3], target = 6
# Output: [0,1]
```

```python
def ind_ad(num):
    for i in range(len(num)):
        if (num[i]+num[i+1])==t:
            return [i,i+1]

num=[3,3]
t=6
ind_ad(num)
```

```python
#Q2.Find the common elements between two lists using dictionary
def find_co(l1,l2):
    d={}
    for i in l1:
        if i in l2:
            d[i]=1
    return d
l1=[1,2,3,4,5]
l2=[3,2,1,5,6]
find_co(l1,l2)
```

```python
#Q5.Write a Binary search Algorithm recursive code
def binarysertree(lst,low,high,key):
    if lst[low]>lst[high]:
        return False
    mid=(low+high)//2
    if lst[mid]==key:
        return True
    elif key<lst[mid]:
        return binarysertree(lst,low,mid-1,key)
    elif lst[mid]<key:
        return binarysertree(lst,mid+1,high,key)
    else:
        return 0
lst=[5,3,7,2,4,6,8]
lst=sorted(lst)
print(binarysertree(lst,0,len(lst)-1,8))
```

```python
#Q6.Merge two sorted List using iterative approach
l1=[4,2,3,4]
l2=[3,4,6,7]
l1.sort()
l2.sort()
print(list(set(l1+l2)))
```

```python
def merg_srt(l1,l2):
    i=j=0
    ans=[]
    while i<len(l1) and j<len(l2):
        if l1[i]<l2[j]:
            ans.append(l1[i])
            i+=1
        else:
            ans.append(l2[j])
            j+=1
    while i<len(l1):
        ans.append(l1[i])
        i+=1
    while j<len(l2):
        ans.append(l2[j])
        j+=1
    return ans
l1=[4,2,3,4]
l2=[3,4,6,7]
merg_srt(l1,l2)
```
