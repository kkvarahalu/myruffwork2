```python
#2. Problem Statement: Given a and b, find prime numbers in a given range [a,b], (a and b areincluded here).
a=int(input())
b=int(input())
for n in range(a,b):  
    is_prime=True  
    for i in range(a,n):  
        if n%i==0:  
            is_prime=False
    if is_prime:
        print(n,end=" ")
    
```

```python
#3. Problem Statement: Check if the given year is a leap year or not.
n=int(input())
if n%4==0 and n%100!=0:
    print("leap year")
    print("YES")
elif n%400==0:
    print("leap year")
    print("YES")
else:
    print("not leap year")
```
