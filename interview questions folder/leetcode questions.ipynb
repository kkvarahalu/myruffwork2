```python
def fun(nums):
    r=nums[0]
    max_end=nums[0]
    for i in range(1,len(nums)):
        max_end=max(nums[i],nums[i]+max_end)
        r=max(r,max_end)
    return r
nums=[5,4,-1,7,8]
fun(nums)
```
