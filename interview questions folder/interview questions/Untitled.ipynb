```python
#fibonic number
def fib(n,a=0,b=1,seq=None):
    if seq is None:
        seq=[]
    elif n<0:
        return seq
    seq.append(a)
    return fib(n-1,b,a+b,seq)
print(fib(10))
```

```python
#frequency number:
def freq(n):
    d={}
    for sr in n:
        if sr in d:
            d[sr]+=1
        else:
            d[sr]=1
    return d
n=[9,8,5,6,7]
print(freq(n))
```

```python
#give all palindrome numbers
def palind(s):
    for i in range(len(s)):
        for j in range(i+1,len(s)+1):
           result=s[i:j]
           if result==result[::-1]:
               print(f"possible palindromes:{result} and lengh {len(result)}")
palind("ababbabbcba")
```

```python
def fun(s):
    m=''
    for i in range(len(s)):
        for j in range(i+1,len(s)+1):
           f=s[i:j]
           if f==f[::-1]:
               if len(f)>len(m):
                   m=f
    return m
print(fun("babad"))
```

```python
# minimum coin problem
coins=[5,3,1] #Descending order
ans=[]
amount=11
for coin in coins:
   while amount >= coin:
       amount-=coin
       ans.append(coin)
print(ans)
print(len(ans))
```
