### q. A person invests ₹40,000 in the year and earns ₹400 interest.in the second year, the investment becomes ₹50000 and the interest earned is ₹500
* identify the pattern or relationship between amount and interest
* derive a formula to calculate interest for any given amount
* predict the interest if the investment amount is ₹60000
* apply linear regression to model this relationship and make predictions
* write python code to automate the prediction using sklearn


### observation:
        400\40000=0.01, 500/50000=0.01===interest =1% of amuont
### formula:
        interest=Amount*0.01
### prediction:
        60000*0.01=₹600
### linear regression:
        x=investiment amount,y=interest        

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
x=np.array([[40000],[50000]])
y=np.array([400,500])
model=lr.fit(x,y)
print(model.predict([[60000]]))
```
