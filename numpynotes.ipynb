```python
def rotetin_matrix(k):
    r,c=len(k),len(k[0])              #this is without predefine functions in numpy print-90 degree matrix
    rotet=[[0]*r for i in range(c)]
    for i in range(r):
       for j in range(c):
           rotet[j][r-1-i]=k[i][j]
    return rotet 
k = [
    [1, 2, 3], 
    [3, 4, 5], 
    [6, 7, 8]
]
rotetin_matrix=rotetin_matrix(k)
for r in rotetin_matrix:
    print(r)
```

```python
# Q7. Write a NumPy program to construct an array by repeating.
# Sample array: [1, 2, 3, 4]
a=np.array([1,2,3,4])
b=[]
for i in a:
   if i not in b:
        b.append(i)
r=np.array(b+b)
r1=np.array(b+b+b)
print(r)
print(r1)
```
